# ETL Pipeline Preparation

### 1. Importing libraries and loading datasets


```python
# Importing required libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import sqlite3
pd.set_option('display.max_colwidth', -1)
```

```python
# loadng messages dataset
messages = pd.read_csv("./data/disaster_messages.csv")
messages.head()
```

```python
# Analyzing unique values of genre
messages['genre'].unique()
```

```python
# Analyzing rows and columns
messages.shape
```

```python
# loading categories dataset
categories = pd.read_csv("./data/disaster_categories.csv")
categories.head()
```

```python
# Analyzing rows and columns
categories.shape
```

### 2. Merging datasets


```python
# Merging categories and messages dataset
df = messages.merge(categories, on='id')
df.head()
```

### 3. Splitting `categories` into separate category columns
As we can see from the above table, categories column (Y) is not in a desired form. All of the categories are clubbed together and we need to separate them.


```python
# Creating a dataframe of the 36 individual category columns
categories = df['categories'].str.split(";", expand=True)
categories.head()
```

```python
# Extracting category names from the first row to label them as column names
row = categories.iloc[0]
category_colnames = row.apply(lambda x: x.split('-')[0])
print(category_colnames)
```

```python
# Renaming the columns of categories
categories.columns = category_colnames
categories.head()
```

### 4. Converting category values to just numbers 0 or 1


```python
for column in categories:
    # setting each value to be the last character of the string
    categories[column] = categories[column].astype(str).str[-1]
    
    # converting column from string to numeric
    categories[column] = categories[column].astype(int)
categories.head()
```

```python
# Checking unique values in columns
for col in categories.columns:
    print(categories[col].unique())
```

We can see that 'related' column has 3 unique values (0, 1 and 2). It is supposed to ideally have 0 or 1 values only because either the message belong to related category or not. So, we need to replace 2 with 1. Also, child_alone column doesn't have any value, so it is better to remove this column.

```python
# Replacing 2 with 1 in 'related' column
categories['related'].replace(2, 1, inplace=True)
```

```python
# Removing 'child_alone' column
categories.drop(['child_alone'], axis=1, inplace=True)
```

### 5. Replacing `categories` column in `df` with new category columns.


```python
# Dropping the original categories column from `df`
df = df.drop(columns=['categories'], axis=1)
df.head()
```

```python
# Concatenating the original dataframe with the new `categories` dataframe
df = pd.concat([df, categories], axis=1)
df.head()
```

### 6. Removing duplicates


```python
# check number of duplicates
df.duplicated().sum() 
```

```python
# Dropping duplicates
df = df.drop_duplicates()
df.head()
```

```python
# Checking number of duplicates again
df.duplicated().sum() 
```

### 6. Saving df into a SQL database


```python
# Creating a SQL database and saving df into a SQL table
engine = create_engine('sqlite:///DisasterResponse.db')
df.to_sql('disaster_messages', engine, index=False)
```

```python
# Connectng to the database
conn = sqlite3.connect('DisasterResponse.db')

# Runing a query
pd.read_sql('SELECT * FROM  disaster_messages', conn)
```
